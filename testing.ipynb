```python
MIN_COLORTEMP = 2200
MAX_COLORTEMP = 6500
TEMP_STEP = 100
SCALING_FACTOR = 1_000_000
from typing import Iterator

total_time=300
time_step=0.3

iterations = round(total_time/time_step)

def get_range(start:int, stop:int, length:int) -> Iterator[int]:
    # multiply then divide by SCALING_FACTOR to avoid issues with `step` rounding to zero.
    # it is possible to calculate the ideal SCALING_FACTOR,
    # but is simpler to just use an arbitrarily large constant
    start = start * SCALING_FACTOR
    stop = stop * SCALING_FACTOR
    span = abs(stop - start)
    step = round((span / length))
    assert step > 0
    iterator = range(start, stop +step, step)
    return (round(x/ SCALING_FACTOR) for x in iterator)

temp_iter = get_range(1, 100, iterations)
len(temp_iter)
```

```python
from bulb_wrapper import Bulb


def rgb_to_hex(rgb: tuple[int, int, int]) -> str:
    return "{:02X}{:02X}{:02X}".format(*rgb)

def hex_to_rgb(hex_colour: str) -> tuple[int, int, int]:
    hex_colour = hex_colour.lstrip('#')
    if len(hex_colour) == 3:
        hex_colour = ''.join([c*2 for c in hex_colour])
    if len(hex_colour) != 6:
        raise ValueError("Hex color must be in the format RRGGBB or RGB")
    r = int(hex_colour[0:2], 16)
    g = int(hex_colour[2:4], 16)
    b = int(hex_colour[4:6], 16)
    return (r, g, b)
```

```python
my_hex = "FF6000"
for t in range(1000, 4000):
    rgb = Bulb.temp_to_rgb(t)
    hex_color = rgb_to_hex(rgb)
    if hex_color == my_hex:
        print(t, rgb, hex_color)
        break
```
